## **2. Loading the Image**
- The image is loaded using `cv2.imread()`. This function reads the image as a NumPy array with three color channels (BGR format by default in OpenCV).

```python
img = cv2.imread("/content/image.jpg")
```

## **3. Checking Image Shape**
- Displays the shape of the image in terms of height, width, and the number of color channels (3 for RGB images).
- Knowing the shape is helpful for image manipulation tasks, including cropping and resizing.

```python
img.shape
```

## **4. Accessing and Printing Pixel Values**
- Specific pixel values at the four corners of the image are accessed and printed. This helps to verify the RGB or BGR values of the pixels, useful for debugging or analyzing pixel data.
- The pixel locations chosen are based on an assumed image size of approximately 580x400. Adjust these values as necessary for images of different dimensions.

```python
print(f"Top left: {img[0, 0]}")
print(f"Top right: {img[0, 400]}")
print(f"Bottom left: {img[580, 0]}")
print(f"Bottom right: {img[580, 400]}")
```

## **5. Splitting Color Channels**
- The `cv2.split()` function is used to separate the three color channels (Blue, Green, and Red) from the original image.
- Each color channel is isolated into a separate 2D array representing intensity values.

```python
blue, green, red = cv2.split(img)
```

## **7. Displaying Color Channel Images**
- Each color channel (blue, green, and red) is displayed as a grayscale image. This is because the single channel only holds intensity values for that color, with higher values indicating more intense color presence.

```python
cv2_imshow(blue)
```

```python
cv2_imshow(green)
```

```python
cv2_imshow(red)
```

## **8. Merging the Color Channels Back**
- The `cv2.merge()` function is used to combine the three color channels back into a single image.
- Displaying `merged_img` verifies that the split and merge operations have been correctly executed, resulting in an image identical to the original.

```python
merged_img = cv2.merge([blue, green, red])
cv2_imshow(merged_img)
```

## **9. Resizing the Image**
- The image is resized to 50% of its original dimensions using `cv2.resize()`, with `INTER_AREA` interpolation for better results when shrinking images.
- Displaying resized_img allows for visualization of the resized output, helpful for applications where image size needs to be standardized.

**Parameters & Variations**:
- `fx=0.5` and `fy=0.5` control the scaling factor for the x and y dimensions. Adjust these values to resize the image by a different factor.
- `interpolation=cv2.INTER_AREA` is ideal for downsizing, while `INTER_LINEAR` or `INTER_CUBIC` may yield better results when upsampling.

```python
resized_img = cv2.resize(img, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
cv2_imshow(resized_img)
```
